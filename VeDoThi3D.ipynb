# Mục mới

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.decomposition import TruncatedSVD, PCA, NMF, SparsePCA, FastICA
from sklearn.preprocessing import MinMaxScaler
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/Python/iris.data.txt', names = ['a', 'b', 'c', 'd', 'e'])
print(df)
labels1 = df['e']
labels2 = pd.factorize(labels1)[0].tolist()
data = df[['a', 'b', 'c', 'd']]
unique_labels = set()

data = PCA(n_components=3).fit_transform(data)
print(data)

data = MinMaxScaler().fit(data).transform(data)
df = pd.DataFrame(data)

colors = ['red', 'green', 'blue']
fig, ax = plt.subplots(figsize=(14, 7), subplot_kw={'projection': '3d'})

for i, row in df.iterrows():
    if labels1[i] not in unique_labels:
        ax.scatter(row[0], row[1], row[2], c=colors[labels2[i]], label=labels1[i])
        unique_labels.add(labels1[i])
    else:
        ax.scatter(row[0], row[1], row[2], c=colors[labels2[i]])

ax.set_xlabel('X Label', fontweight ='bold')
ax.set_ylabel('Y Label', fontweight ='bold')
ax.set_zlabel('Z Label', fontweight ='bold')
ax.legend(loc='upper left', title='Name item Iris', fontsize=12, title_fontsize=14)
plt.show()

```
